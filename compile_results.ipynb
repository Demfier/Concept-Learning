```python
"""
This notebook combines the result obtained on the dataset
"""

import os
import json
import _pickle as pickle
import pandas as pd
from IPython.display import display

RE_EVAL_DIR = 'data/out/eval/'
PAC_DIR = 'data/out/pac/'
RE_DIR = 'data/out/regex/'

re_uncov_results = os.listdir(RE_EVAL_DIR + 'uncov/')

results = []
for result_file in re_uncov_results:
    result = json.load(open(RE_EVAL_DIR + 'uncov/' + result_file, 'r'))
    result_df = pd.DataFrame()
    result_df['language'] = result.keys()
    result_df['accuracy'] = [value[0] for value in result.values()]
    results.append(result_df)
print("Uncovered evaluation result on `medium` data")
results[0].sort_values(by=['accuracy'], ascending=False).reset_index()
```

```python
print("Uncovered evaluation result on `high` data")
results[1].sort_values(by=['accuracy'], ascending=False).reset_index()
```
