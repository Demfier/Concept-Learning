```python
"""
This notebook combines the result obtained on the dataset
"""

import os
import json
import _pickle as pickle
import pandas as pd
from IPython.display import display

RE_EVAL_DIR = 'data/out/eval/'
PAC_DIR = 'data/out/pac/'
RE_DIR = 'data/out/regex/'

out_files = os.listdir(PAC_DIR)

lang = []
accuracies = []
opn_seq = []
for file in out_files:
    pac_data = pickle.load(open(PAC_DIR + file, 'rb'))
    try:
        opn_seq.append(pac_data[3])
    except IndexError:
        continue
    lang.append(file.strip('.p'))
    accuracies.append(pac_data[0])
result_df = pd.DataFrame()
result_df['language'] = lang
result_df['accuracy (common words)'] = accuracies
result_df['opn seq eg'] = opn_seq
result_df.sort_values(by='accuracy (common words)', ascending=False)
```
