```python
#Load Corpus
f = open('polish-train-high').read().splitlines()
```

```python

import pandas

from collections import defaultdict


# Preprocessing the shared task data to a JSON/Dictionary
fk = defaultdict(dict)
inti = 0
for item in f:
    stuff = item.split('\t')
    fk[inti]['source'] = stuff[0]
    fk[inti]['target'] = stuff[1]
    try:
        fk[inti]['type'],fk[inti]['subtype'] = stuff[2].split(';')[0].split('.')
    except:
        fk[inti]['type'] = stuff[2].split(';')[0]
        
    fk[inti]['props'] = stuff[2].split(';')[1:]
    inti += 1

# Convert a table to Dataframe
pd = pandas.DataFrame.from_dict(fk,orient='index')

# Gramamtical categories as separate columns p0 to p4
pd = pd.join(pandas.DataFrame(pd.props.values.tolist()))
pd.columns = ['source', 'target', 'type', 'props', 'subtype', 'p0', 'p1', 'p2', 'p3', 'p4']
pd.head()
```

```python
import string
import re

# Iterative longest contiguous sequence. No one character matchings

def lcs(s1,s2):
    longest = ""
    i = 0
    for x in s1:
        if re.search(x, s2):
          s = x
          while re.search(s, s2):
            if len(s)>len(longest):
                longest = s
            if i+len(s) == len(s1):
                break
            s = s1[i:i+len(s)+1]
        i += 1
    return longest

def iterLCS(pdf):
    sw1 = pdf['source']
    sw2 = pdf['target']
    longList = []
    while True:
        tempVal = lcs(sw1,sw2)
        if len(tempVal)  <= 1:
            break
        
        longList.append(tempVal)
        sw1 = sw1.replace(tempVal,'#',1)
        sw2 = sw2.replace(tempVal,'!',1)
    #return [longList,[item for item in sw1.split('#') if len(item) > 0],[item for item in sw2.split('!') if len(item) > 0]]
    pdf['common'] = longList
    pdf['deleted'] = [item for item in sw1.split('#') if len(item) > 0]
    
    if len(pdf['deleted']) == 0:
        pdf['deleted'] = ['ϵ']
    
    pdf['added'] = [item for item in sw2.split('!') if len(item) > 0]
    
    if len(pdf['added']) == 0:
        pdf['added'] = ['ϵ']
    
    return pdf
    
#iterLCS(newPD2.ix[3,:])
```

```python
%timeit
newPD2 = pd.apply(iterLCS, axis=1)
newPD2.head()
```

```python
#Final Table
b= pandas.DataFrame(newPD2['added'].tolist(),index=newPD2.index).stack().reset_index()
b.columns = ['mainInd','subInd','addedSub']
fin = pandas.DataFrame.merge(b,newPD2,left_on='mainInd',right_index=True,how='left')
fin['addObj'] = ''
for i,item in fin.iterrows():
    fin.ix[i,'addObj'] = fin.ix[i,'added'][fin.ix[i,'subInd']]
fin.head(10)
```

```python
def lcs(SB,TB):
    S = SB
    T = TB
    m = len(S)
    n = len(T)
    
    
    globList = list()
    
    while True:
        lcs_set = set()
        lcsList = list()
        counter = [[0]*(n+1) for x in range(m+1)]
        longest = 0
        for i in range(m):
            for j in range(n):
                if S[i] == T[j]:
                    c = counter[i][j] + 1
                    counter[i+1][j+1] = c
                    if c > longest:
                        lcs_set = set()
                        longest = c
                        lcs_set.add(S[i-c+1:i+1])
                    elif c == longest:
                        lcs_set.add(S[i-c+1:i+1])
        
        lcsList = list(lcs_set)    
        globList.extend(lcsList)
        if len(lcs_set) == 0:
            break
        for item in lcsList:
            S = S.replace(item,'_',1)
            T = T.replace(item,'$',1)
            m = len(S)
            n = len(T)
    
    return [item for item in globList if len(item) > 1]    
```
